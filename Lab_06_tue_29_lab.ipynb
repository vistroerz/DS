1. #3 (age): age in years 
2. #4 (sex): sex (1 = male; 0 = female) 
3. #9 (cp): cp: chest pain type | Value 0: typical angina | Value 1: atypical angina | Value 2: non-anginal pain | Value 3: asymptomatic 
4. #10 (trestbps): resting blood pressure (in mm Hg on admission to the hospital) 
5. #12 (chol): serum cholestoral in mg/dl 
6. #16 (fbs): (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false) 
7. #19 (restecg): resting electrocardiographic results | Value 0: normal | Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) | Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria 
8. #32 (thalach): maximum heart rate achieved 
9. #38 (exang): exercise induced angina (1 = yes; 0 = no) 
10. #40 (oldpeak): ST depression induced by exercise relative to rest 
11. #41 (slope): the slope of the peak exercise ST segment | Value 1: upsloping | Value 2: flat | Value 3: downsloping 
12. #44 (ca): number of major vessels (0-3) colored by flourosopy 
13. #51 (thal): 3 = normal; 6 = fixed defect; 7 = reversable defect 
14. #58 (num) (the predicted attribute): Value 0: < 50% diameter narrowing | Value 1: > 50% diameter narrowing

```python
X = df.drop('target', axis=1)
# target isimli kolon haricini X e atamak
```

```python
y = df.target
# y değerini atamak
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# data setinin split yapılması
```

```python
scaler = MinMaxScaler()
# ** eğer  sütunlar sadece 0-1 lerden oluşuyor ise(cinsiyet gibi)  o zaman standardScale olmaz
# ** hem categorik hem booling değerler olan veri setinde StandartScaler kullanılmaz
# hangi veri üzerinde hangi scaler kullanmamız gerektiğini iyi bilmeliyiz.Piyasadakiler ezbere yapıyor
```

```python
# bazen de 0-1 olan sütunlar ayırt edilip onlara MinMax scaling, diğer contin. olanlara standart scaling ....
# ... yapılıyor ve sonra pandas ile concat yapılıyor bu veri seti.. 
```

```python
X_train_scaled = scaler.fit_transform(X_train)
# Bu notebook da hem scele gerekli olan hem de gerekli olmayan modeller için ayrı ayrı \
# \ scale edilmiş varıable lar oluşturuluyor
```

```python
X_test_scaled = scaler.transform(X_test)
# hem scle edilen hem edilmeyen veriler oluşturduk
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train_scaled, y_train)
# KNN için scaled olan veri kullanılıyor
```

```python
knn_pred = knn.predict(X_test_scaled)
```

```python
print(classification_report(y_test, knn_pred))
# 1 için recall değerine baktık..çok düşük..0.76..yani hastaların %25 ini kaybedebiliriz.
```

```python
sns.heatmap(confusion_matrix(y_test, knn_pred), annot=True)
# 12 hasataya kalp krizi geçirmez demiş model ama hatalı değerlendirme olmuş..
```

```python
cm = ConfusionMatrix(knn)
cm.fit(X_train_scaled, y_train)
cm.score(X_test_scaled, y_test)
cm.show()
# üstteki gösterimin aynısı ama burada true ve predicted olan alanları otomatik olarak yazıyor
```

```python
# elbow metodu uygulaması

error_rate = []# Will take some time
for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train_scaled,y_train)
    pred_i = knn.predict(X_test_scaled)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')

# alttaki grafiğe göre k valu e olarak 5 - 9 seçebiliriz. k yı mümkün olarak düşük seçeceğiz. 5' i seçiyoruz.
# n_neighbors=5 seçeceğiz
```

```python
knn.fit(X_train_scaled, y_train)
```

```python
print(classification_report(y_test, knn_pred))

# 1 için recall arttı, 0,80 oldu. ama halen hastaların %20 sini kalp krizi geçirmeyecek diye düşünüp \\ 
# \\ yanlış bir tahmin yapıyoruz.
```
